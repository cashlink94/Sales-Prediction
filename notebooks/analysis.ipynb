# Sales Data Exploratory Analysis
This notebook explores the extended sales dataset to understand trends and patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/sales_prediction_data_extended.csv', parse_dates=['Date'])
df.head()
```

## Summary Statistics

```python
df.describe()
```

```python
df.info()
```

## Sales Trend Over Time

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Sales'], color='blue')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## Average Sales per Weekday

```python
weekday_avg = df.groupby(df['Date'].dt.dayofweek)['Sales'].mean()
plt.figure(figsize=(8,5))
sns.barplot(x=weekday_avg.index, y=weekday_avg.values, palette='viridis')
plt.xticks(ticks=range(7), labels=['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.title('Average Sales per Weekday')
plt.ylabel('Average Sales')
plt.show()
```

## Observations
- Sales show weekly seasonality.
- Weekends tend to have lower sales.
- Overall trend shows gradual increase due to extended dataset simulation.
